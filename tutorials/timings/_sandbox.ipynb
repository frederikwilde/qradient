```python
variances = []
var_mins = []
var_maxs = []
snum = 100
qnums = np.arange(2, 11, 2)
for qnum in qnums:
    print(qnum, end='\r')
    sys.stdout.flush()
    lnum = 100
    zz = np.full((qnum, qnum), None)
    zz[0, 1] = 1.
    crct = qradient.circuit_logic.McClean(qnum, {'zz': zz}, lnum)
    axes = (3 * np.random.rand(snum, lnum, qnum)).astype('int')
    angles = 2*np.pi * np.random.rand(snum, lnum, qnum)
    expectations = []
    for i in tnrange(snum):
        crct.axes = axes[i]
        crct.angles = angles[i]
        _, g = crct.grad_run(hide_progbar=True)
        expectations.append(g[0,0])
    v = stats.bayes_mvs(expectations)[1]
    variances.append(v[0])
    var_mins.append(v[1][0])
    var_maxs.append(v[1][1])
```

```python
reg = stats.linregress(qnums, np.log(np.array(variances)))
print("Slope = {}".format(reg[0]))
abscissa = np.arange(min(qnums), max(qnums)+.1, .1)
plt.plot(abscissa, np.exp(reg[0]*abscissa + np.full(len(abscissa), reg[1])), 'r--')
plt.errorbar(
            qnums,
            variances,
            yerr=[var_mins, var_maxs],
            marker="."
)
plt.semilogy()
plt.show()
```

```python
def plot_opt(costs, parameters):
    plt.figure(figsize=(15, 2))
    gs = gridspec.GridSpec(1, 2, width_ratios=[1, 4]) 
    plt.subplot(gs[0])
    plt.semilogy()
    plt.plot(np.array(costs) + 1, '-')
    plt.subplot(gs[1])
    plt.plot(parameters, '-')
    plt.show()
```

```python
qnum, lnum = 5, 5
c = McClean(qnum, None, lnum)
# observable
z = sp.csr_matrix([[1., 0.], [0., -1.]], dtype='complex')
zz = sp.kron(z, z)
zz = sp.kron(zz, sp.identity(2**(qnum-2)), format='csr') + \
    sp.kron(sp.identity(2), sp.kron(zz, sp.identity(2**(qnum-3))), format='csr') + \
    sp.kron(z, sp.kron(sp.identity(2), sp.kron(z, sp.identity(2**(qnum-3)))), format='csr')
c.observable = zz
# initial guess
#init = 2*np.arctan(.5*np.random.normal(0, .01, [lnum, qnum]))
init = np.pi * np.random.rand(lnum, qnum)
plt.figure(figsize=(4, 1.5))
plt.hist(init.flatten(), bins=np.arange(-np.pi/2, np.pi/2, .2))
plt.show()
```

```python
for r in [1]:
    c.noise = 0.
    c.regularize = r
    steps = 100
    rate = 1
    costs, parameters, diffs = np.zeros(steps+1), np.zeros([steps+1, lnum*qnum]), np.zeros([steps, lnum*qnum])
    c.angles = init
    costs[0], parameters[0] = c.expec_val().real, init.flatten()

    for i in tnrange(steps):
        c.regularize = r/(i+1)
        costs[i+1], p = c.opt_step(rate)
        parameters[i+1] = p.flatten()
        diffs[i] = parameters[i+1] - parameters[i]
    print('cost={}, noise={}, regularize={}'.format(costs[-1], c.noise, r))
    plot_opt(costs, parameters)
```
